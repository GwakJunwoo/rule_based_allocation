```python
import pandas as pd
import pandas_datareader.data as pdr
import datetime

# 티커 목록 읽기
codes = pd.read_csv('tickers.csv')
tickers = codes['ticker'].tolist()  # 'ticker' 컬럼의 모든 값들을 리스트로 변환

# 시작 날짜 설정
start_date = datetime.datetime(2010, 1, 1)

# 수정 종가를 저장할 빈 데이터프레임 생성
adj_close_prices = pd.DataFrame()

# 각 티커에 대해 데이터 가져오기
for ticker in tickers:
    try:
        data = pdr.get_data_yahoo(ticker, start_date)
        adj_close_prices[ticker] = data['Adj Close']
    except Exception as e:
        print(f"Error retrieving data for {ticker}: {e}")

# 최종 데이터프레임 반환
adj_close_prices.head()  # 예시로 처음 몇 줄만 보여줍니다.
```

```python
adj_close_prices.to_csv('dataset.csv')
```
